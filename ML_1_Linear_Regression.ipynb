```python
# Import libraries (we'll use these to work with data and make plots)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Create a simple dataset: ad spend and sales for 100 stores
np.random.seed(42)  # For reproducibility
ad_spend = np.random.uniform(1000, 5000, 100)  # Ad spend between $1,000 and $5,000
sales = 20000 + 5 * ad_spend + np.random.normal(0, 2000, 100)  # Sales = 20000 + 5 * ad_spend + noise
data = pd.DataFrame({'Ad_Spend_USD': ad_spend, 'Sales_USD': sales})

```

```python
# Show the first few rows of the data
print("Here’s our data: Ad Spend and Sales for 100 stores")
print(data.head())
```

```python
# Plot the data as a scatter plot
plt.scatter(data['Ad_Spend_USD'], data['Sales_USD'], color='blue', label='Data')
plt.xlabel('Ad Spend (USD)')
plt.ylabel('Sales (USD)')
plt.title('Ad Spend vs Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Prepare the data for machine learning
X = data[['Ad_Spend_USD']]  # Input (ad spend)
y = data['Sales_USD']       # Output (sales)

# Create and train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
predictions = model.predict(X)
```

```python
# Plot the data with the regression line
plt.scatter(data['Ad_Spend_USD'], data['Sales_USD'], color='blue', label='Data')
plt.plot(data['Ad_Spend_USD'], predictions, color='red', label='Linear Fit')
plt.xlabel('Ad Spend (USD)')
plt.ylabel('Sales (USD)')
plt.title('Linear Regression: Predicting Sales from Ad Spend')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Print the slope and intercept
print(f"Slope (Ad Spend impact): ${model.coef_[0]:.2f} per $1 ad spend")
print(f"Intercept (Base Sales): ${model.intercept_:.2f}")
r_squared = model.score(X, y)
print(f"R-squared: {r_squared}")
```

```python
# Predict sales for a new ad spend
new_ad_spend = 1  # Try changing this value!
predicted_sales = model.predict([[new_ad_spend]])
print(f"If we spend ${new_ad_spend} on ads, predicted sales are ${predicted_sales[0]:.2f}")
```
